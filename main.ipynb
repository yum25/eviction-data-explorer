```python
import altair as alt
import pandas as pd
import requests
import io
```

```python
s3_eviction_lab_base = "https://data-downloads.evictionlab.org/"
data_for_analysis = "data-for-analysis/"
data_court_issued = "county_court-issued_2000_2018.csv"
demographics = "demographics-of-eviction-data/"

legacy_data_MI_tract = "legacy-data/validated/tracts/MI_tracts.csv"

def defineS3Path(paths:list[str]):
    return ''.join([s3_eviction_lab_base, *paths])
```

```python
response = requests.get(defineS3Path([legacy_data_MI_tract]))

if response.status_code == 200:
     df = pd.read_csv(io.StringIO(response.text))
     print(df)
else:
    print(f"Request failed with status code: {response.status_code}")
```
